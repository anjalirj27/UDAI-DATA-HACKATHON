```python
# 1. Feature Engineering on Cleaned Data
# -------------------------------------

# Total activity
df_cleaned['total_activity'] = (
    df_cleaned['demo_age_5_17'] + df_cleaned['demo_age_17_']
)

# Youth ratio (%)
df_cleaned['youth_ratio'] = (
    df_cleaned['demo_age_5_17'] / (df_cleaned['total_activity'] + 1e-6)
) * 100

# Convert date (DD-MM-YYYY) to datetime
df_cleaned['date'] = pd.to_datetime(
    df_cleaned['date'],
    format='%d-%m-%Y'
)

# Extract time features
df_cleaned['month'] = df_cleaned['date'].dt.month_name()
df_cleaned['day_of_week'] = df_cleaned['date'].dt.day_name()
```

```python
# Per-capita activity (proxy using total_activity itself as denominator base)

df_cleaned['activity_per_capita'] = (
    df_cleaned['total_activity'] /
    (df_cleaned['total_activity'].mean() + 1e-6)
)
```

```python
df_cleaned.head()
```

```python
# Create urban-rural proxy using quantiles
df_cleaned['urban_rural_proxy'] = pd.qcut(
    df_cleaned['total_activity'],
    q=3,
    labels=['Rural-like', 'Semi-Urban', 'Urban-like']
)
```

**ML Clustering of Districts (K-Means)**

🔹 Step 1: District-level dataset

```python
district_ml = (
    df_cleaned
    .groupby('district')
    .agg(
        avg_youth_ratio=('youth_ratio', 'mean'),
        avg_activity=('activity_per_capita', 'mean'),
        total_activity=('total_activity', 'sum')
    )
    .reset_index()
)
```

🔹 Step 2: Scaling

```python
from sklearn.preprocessing import StandardScaler

features = district_ml[['avg_youth_ratio', 'avg_activity']]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(features)
```

🔹 Step 3: K-Means Clustering

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)
district_ml['cluster'] = kmeans.fit_predict(X_scaled)
```

🔹 Step 4: Cluster interpretation

```python


district_ml.groupby('cluster')[['avg_youth_ratio', 'avg_activity', 'total_activity']].mean()
```

Cluster 0 → Urban, high activity, moderate youth

Cluster 1 → Rural, low activity, high youth

Cluster 2 → Transitional districts

```python
# --- STEP 6: ANOMALY DETECTION (Finding Spikes) ---

# 1. Calculate daily total activity
daily_stats = df_cleaned.groupby('date')['total_activity'].sum().reset_index()

# 2. Compute mean and standard deviation of daily activity
mean_val = daily_stats['total_activity'].mean()
std_val = daily_stats['total_activity'].std()

# 3. Set anomaly threshold (Mean + 2 * Standard Deviation)
# Any value above this threshold is considered an "anomaly"
threshold = mean_val + 2 * std_val
anomalies = daily_stats[daily_stats['total_activity'] > threshold]

print("--- Anomaly Detection Report ---")
if len(anomalies) > 0:
    print(f"Detected {len(anomalies)} anomalous days with unusually high activity:")
    print(anomalies)
else:
    print("No major spikes detected. Activity is stable.")

# Plot anomalies
plt.figure(figsize=(12,6))
plt.plot(
    daily_stats['date'],
    daily_stats['total_activity'],
    label='Daily Activity',
    color='blue',
    alpha=0.5
)
plt.scatter(
    anomalies['date'],
    anomalies['total_activity'],
    color='red',
    label='Anomalies (Spikes)',
    zorder=5
)
plt.axhline(
    threshold,
    color='green',
    linestyle='--',
    label='Anomaly Threshold'
)
plt.title('Daily Aadhaar Activity & Anomaly Spikes', fontsize=15)
plt.legend()
plt.savefig('anomaly_detection.png')
plt.show()
```

```python
# --- STEP 7: PINCODE HOTSPOT ANALYSIS ---

# 1. Calculate total activity at the district level
dist_totals = df_cleaned.groupby(['state', 'district'])['total_activity'].sum().reset_index()
dist_totals.columns = ['state', 'district', 'dist_total']

# 2. Merge pincode-level activity with district totals
pincode_hubs = df_cleaned.groupby(['state', 'district', 'pincode'])['total_activity'].sum().reset_index()
hotspots = pd.merge(pincode_hubs, dist_totals, on=['state', 'district'])

# 3. Calculate percentage contribution of each pincode to its district
hotspots['contribution_pct'] = (hotspots['total_activity'] / hotspots['dist_total']) * 100

# 4. Identify top 10 "Monopoly" pincodes
# (Pincodes where a single pincode handles more than 70% of the district's total activity)
monopolies = hotspots[hotspots['contribution_pct'] > 70] \
    .sort_values(by='total_activity', ascending=False)

print("\n--- Service Delivery Hotspots ---")
print(monopolies[['state', 'district', 'pincode', 'contribution_pct']].head(10))
```
